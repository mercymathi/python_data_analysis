```python
def format_phone_number(num):
    if len(num) == 10:  # Check if it's exactly 10 digits
        return f"{num[0:3]}-{num[3:6]}-{num[6:10]}"
    return num  # Return as is if not 10 digits

# Apply formatting to the column
df["phone_number"] = df["phone_number"].apply(format_phone_number)
```
